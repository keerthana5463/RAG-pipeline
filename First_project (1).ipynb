Retrieve relevant documents for a query

```python
retriever = vectorstore.as_retriever()
```

```python
!pip install langchain-groq
```

```python
GROQALEX='yourapikey'
```

Generate answers using an LLM with retrieved context

```python
from langchain_groq import ChatGroq
#os.environ['groq_api_key'] = userdata.get('GROQALEX')
llm=ChatGroq(
    model="openai/gpt-oss-20b",
    groq_api_key=GROQALEX,
    temperature=0.3,
    max_tokens=200
)
```

```python
from langchain_classic.chains import RetrievalQA

qa_chain=RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=True
)
```

```python
query1="what does Awareness of Sound?"
response1 = qa_chain.invoke({'query':query1})

print("Query:",query1)
print("Answer:",response1["result"])
print("\nSources:",[doc.metadata for doc in response1["source_documents"]])
```
